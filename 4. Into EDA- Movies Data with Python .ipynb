## What is EDA
### EDA is for understanding dataset.
### Brfore getting into formal analysis of data, it is good to investigate the dataset- this investigation is nothing but EDA

## Benefits of EDA
### It provides a brief summary of data on which we are going to work, also provides better sense of data along with some statistics & visualization. 

### NOTE: Here we are going to study each variable separately, which is often called "univariate summaris".  Here featurs are "quantative"(numeric)  or "categorical"(discrete). We can use summary accordingly.

### Not necessarly but sometimes we do EDA to find interesting insights from datasets.


```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# imort data kaggle link: https://www.kaggle.com/datasets/paradisejoy/top-hits-spotify-from-20002019
df =pd.read_csv('songs_spotify.csv')
df.head() # shows first five rows
```

```python
df.tail() # shows last five rows
```

```python
# Check data type for each column
df.dtypes
```

```python
df.describe() # shows data stats
```

```python
df.shape # shows data have 2000 rowa & 18 columns
```

```python
# since the dataset is provided with songs from 2000 to 2020, hence lets check unique year counts
df.year.unique()
# here we have some songs from year 1999, 1998 & 2020 
```

```python
# lets check howmany songs are there from year 1998, 1999 & 2020 before deleting it(the entire rows)
print(len(df.query('year==1998')))
print(len(df.query('year==1999')))
print(len(df.query('year==2020')))

# so we have 1, 38 & 3 songs from year 1998, 1999 & 2020
```

```python
# now lets remove the data from year 1998, 1999 & 2020
# step1: lets store the data to be deleated in a new data frame 
df_to_delete = df[(df['year'] <2000) | (df['year'] >2019)].index
# Step2: drop or remove data & store it in original dataframe df
df = df.drop(df_to_delete)
# Check again if years dropped.
df.year.unique()
# now its dropped
```

```python
df
# now the no of rows reduced from 2000 to 1958 
```

```python
# Now lets check for any duplicate rows
df_duplicate_rows = df[df.duplicated()]
print("number of duplicate rows present: ", df_duplicate_rows.shape)
df_duplicate_rows.head(59)

# To drop a column we use df = df.drop(['year', 'popularity'], axis=1) similarly we can also drop rows(axis=0)
# To rename columns use df = df.rename(columns={"artist": "artist_name", "song": "song_lyrics" }) we may add multiple columns
```

```python
#before deleting lets count the number of rows
df.count() 
```

```python
# Lets Delete the duplicate rows and count again
df = df.drop_duplicates()
df.count()
# now the count is 59 less(the duplicate values)
```

```python
# verify for any duplicate value:
df.duplicated().sum()
```

```python
# Finding null values in dataset if any
df.isnull().sum()
# Good, no null values found, if any found we use 
# df.replace(np.nan, 0, inplace=True) to replace null values to zero(or anyvalue)

# use df = df.dropna()    to Drop the missing values.
```

```python
# lets check no of songs released per year
songs_year_wise = (df.year.value_counts().rename_axis('year').reset_index(name='songs'))
songs_year_wise
```

```python
# plot songs per year 
plt.figure(figsize=(14,8))
ax = sns.barplot(x = 'year', y = 'songs', data = songs_year_wise, palette='deep')
plt.ylabel('Number of songs', size = 15)
plt.xlabel('Year', size = 15)
plt.title('Number of songs released per year', size = 20)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15, rotation = 50)
#ax.bar_label(ax.containers[0],size = 15)
plt.show()

# here we see that 2012 has maximum no of songs released, whie in 2000 lowest release was there.
```

```python
# lets analyse explicit content with dataframe
explicit_or_not_explicit = pd.DataFrame(df.explicit.value_counts().rename_axis(
    'explicit').reset_index(name='songs'))
explicit_or_not_explicit

# here only 526 songs were explicit
```

```python
df.corr()
sns.heatmap(df.corr())
```

```python
# Most popular artists
# Organizing the ten most popular artists according to popularity score
top_10_artists = df.groupby('artist')[['artist','popularity']].sum().sort_values('popularity',
                                                               ascending=False).head(10)
top_10_artists
```

```python
# Visualizing most popular artists
plt.figure(figsize=(14,8))
ax = sns.barplot(data = top_10_artists, x = 'popularity', y = top_10_artists.index ,palette='deep')
plt.ylabel('Artist', size = 15)
plt.xlabel('Tota popularity', size = 15)
plt.title('Most popular artists from 2000-2019 per popularity score', size = 20)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15, rotation = 50)
# ax.bar_label(ax.containers[0],size = 15)
plt.show()
```
